```python
from plotnine import *
import pandas as pd
```

```python
df = pd.read_csv('Jordan\\Energy.csv')
df
```

```python
df = df.drop(df.index[-1])
df
```

```python
df['StartTime'] = pd.to_datetime(df['StartTime'], format='%H:%M')
df
```

```python
(
ggplot(df) + 
    geom_line(aes(x=df.index, y='Generation'), color='blue', linetype='solid', size=1) +
        geom_line(aes(x=df.index, y='Load'), color='red', linetype='dashed', size=1) +
    labs(title = "Generation of Energy by Time",
         x = "Time Since Midnight (Hours)",
         y = "Energy (kwph)")+
    theme_bw() 
)
```
