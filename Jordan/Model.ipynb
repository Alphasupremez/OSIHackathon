# This code will show how the model works

```python
import pandas as pd
```

```python
Energy = pd.read_excel("Energy.xlsx")
Energy
```

```python
Energy['Gen_Load'] = Energy['Generation'] - Energy['Load']
Energy
```

```python
base_rate = .60
low_rate = .50
med_rate = .40
high_rate = .30
```

```python
columns = ["StartTime","EndTime", "Rate"]
save_time = pd.DataFrame(columns = columns)
save_time
```

```python
for index, rows in Energy.iterrows():
    start_time = rows['StartTime']
    end_time = rows['EndTime']
    genlvl = rows['GenLvl']
    genload = rows['Gen_Load']
    if genload > 0:
        if genlvl == 'low':
            rate = 0.5
        elif genlvl == 'med':
            rate = 0.4
        elif genlvl == 'high':
            rate = 0.3
    else:
        rate = 0.6
    save_row = pd.Series({'StartTime': start_time, 'EndTime': end_time, 'Rate': rate})
    save_time = save_time.append(save_row, ignore_index = True)
```

```python
save_time.to_csv("save_time.csv")
```
